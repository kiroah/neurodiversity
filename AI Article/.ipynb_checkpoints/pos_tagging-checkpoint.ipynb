```python
from ipywidgets import interact, IntSlider
from IPython.display import display

page_size = 10

df = words_df.sort_values(by=0)

def show_page(page_number):
    start_index = (page_number - 1) * page_size
    end_index = page_number * page_size
    page_df = df.iloc[start_index:end_index]
    display(page_df)

# Create an interactive widget to page through the DataFrame
interact(show_page, page_number=IntSlider(min=1, max=len(df) // page_size + 1, step=1, value=1))
```
